```python
from __future__ import print_function
import fibre
import ref_tool
# logger verbose=True
logger = fibre.utils.Logger(verbose=True)
# Find an REF-Unit that is connected on the serial port /dev/ttyUSB0
# my_drive = ref_tool.find_any("serial:/dev/ttyUSB0")
teach_arm = ref_tool.find_any(serial_number="3950366E3233", logger=logger)
```

```python
from __future__ import print_function
import fibre
import ref_tool
# logger verbose=True
logger = fibre.utils.Logger(verbose=True)
# Find an REF-Unit that is connected on the serial port /dev/ttyUSB0
follow_arm = ref_tool.find_any(serial_number="395D36713233", logger=logger)
```

```python
# Enable motor but disable circle udpate
teach_arm.robot.joint_all.set_enable(True)
teach_arm.robot.set_enable(False)
```

```python
# Enable motor but disable circle udpate
follow_arm.robot.joint_all.set_enable(True)
follow_arm.robot.set_enable(False)
```
